```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

plt.style.use('ggplot')
```

```python
X, Y = make_blobs(n_samples=500, centers=3)
```

```python
plt.scatter(X.T[0], X.T[1]);
```

```python
model = KMeans(n_clusters=3)
```

```python
pred = model.fit_predict(X)
```

```python
plt.scatter(X.T[0], X.T[1], c=pred);
```

```python
cluster = X[np.where(pred == 1)]
```

```python
plt.scatter(cluster.T[0], cluster.T[1]);
```

This is the green cluster above.

```python
pca = PCA(n_components=1)
```

```python
pca.fit_transform(cluster);
```

```python
# from https://stackoverflow.com/questions/22984335/recovering-features-names-of-explained-variance-ratio-in-pca-with-sklearn
print(pd.DataFrame(pca.components_, columns=['col1', 'col2'], index=['PC-1']))
```

This is what we want! According to this, the first column (x-axis) is more important, which is exactly what is seen in the first graph.
